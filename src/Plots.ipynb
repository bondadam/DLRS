```python
import pandas as pd
import seaborn as sns
```

```python
loc_loss = pd.read_csv("od_localisation_loss.csv")
cla_loss = pd.read_csv("od_classification_loss.csv")
```

```python
loc_loss
```

```python
g = sns.lineplot(data=loc_loss, x="Step", y="Value")
g.set_title("Localisation Loss for the Anomaly Object Detection Model")
g.set_xlabel("Iteration")
g.set_ylabel("Localisation Loss")
```

```python
g = sns.lineplot(data=cla_loss, x="Step", y="Value")
g.set_title("Classification Loss for the Anomaly Object Detection Model")
g.set_xlabel("Iteration")
g.set_ylabel("Classification Loss")
```
